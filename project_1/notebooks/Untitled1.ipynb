```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

DATPATH = "../datafiles/"
```

```python
# Load the terrain
terrain1 = imread(DATPATH + "SRTM_data_Norway_1.tif")
print(len(terrain1))

terrain1_low_res = terrain1[::20, ::20]
print(len(terrain1_low_res))

# Show the terrain
plt.figure()
plt.subplot(121)
plt.title("High Resolution")
plt.imshow(terrain1, cmap="gray")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(122)
plt.imshow(terrain1_low_res, cmap="gray")
plt.title("Low Resolution")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
