```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams.update(
    {
        "text.usetex": True,
        "font.family": "serif",
        # "font.sans-serif": ["Helvetica"]
    }
)
```

```python
tn = 1
dx = 0.1
dt = 0.1

x = np.linspace(0, 1, int(1 / dx))
t = np.linspace(0, tn, int(tn / dt))

X, T = np.meshgrid(x, t)
Z = u(X, T)

fig, ax = plt.subplots(figsize=(4, 3))
cf = ax.pcolormesh(X, T, Z, cmap=plt.get_cmap("inferno"))
fig.colorbar(cf, ax=ax)
ax.set_xlabel("$x$")
ax.set_ylabel("$t$")
#ax.set_rasterized(True)
```

```python
def u(x, t):
    return np.sin(np.pi * x) * np.exp(- np.pi * t)

tn = 1
dx = 0.01
dt = 0.01

x = np.linspace(0, 1, int(1 / dx))
t = np.linspace(0, tn, int(tn / dt))

X, T = np.meshgrid(x, t)
Z = u(X, T)

fig, ax = plt.subplots(figsize=(4, 3))
cf = ax.pcolormesh(X, T, Z, cmap=plt.get_cmap("inferno"))
fig.colorbar(cf, ax=ax)
ax.set_xlabel("$x$")
ax.set_ylabel("$t$")
#ax.set_rasterized(True)
plt.savefig("../figs/heat_eqn_analytic.pdf")
```
